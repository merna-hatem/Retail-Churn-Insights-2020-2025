```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Retail.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['WarehouseLocation'].fillna("Unknown",inplace=True)
```

```python
df["Total Sales"]=df["Quantity"]*df["UnitPrice"]
```

```python
df=df[(df['Quantity']>0) & (df['UnitPrice']>0)]
```

```python
LastOrderDate=max(df['InvoiceDate'])
recency=df.groupby('CustomerID')['InvoiceDate'].max().reset_index()
recency['Recency']=(LastOrderDate-recency['InvoiceDate']).dt.days
```

```python
frequency=df.groupby('CustomerID')['InvoiceNo'].nunique().reset_index()
frequency.rename(columns={'InvoiceNo':'Frequency'}, inplace=True )
```

```python
monetary=df.groupby('CustomerID')['Total Sales'].sum().reset_index()
monetary.rename(columns={'Total Sales':'Monetary'}, inplace=True )
```

```python
rfm=recency.merge(frequency, on='CustomerID').merge(monetary, on='CustomerID')
```

```python
rfm.head()
```

```python
rfm['Churn Flag']=np.where(rfm['Recency']<90,'Active','Churned')
```

```python
rfm.head(15)
```

```python
rfm['R_Score']=pd.qcut(rfm['Recency'],5, labels=[5,4,3,2,1])
rfm['F_Score'] = pd.cut(
    rfm['Frequency'],
    bins=[0,1,2,3,5,100],
    labels=[1,2,3,4,5])
rfm['M_Score']=pd.qcut(rfm['Monetary'],5, labels=[1,2,3,4,5])
rfm['RFM_Score']=rfm['R_Score'].astype(str)+rfm['F_Score'].astype(str)+rfm['M_Score'].astype(str)
```

```python
rfm.head(10)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18,5))

sns.histplot(rfm['Recency'], bins=30, ax=axes[0])
axes[0].set_title('Recency')

sns.histplot(rfm['Frequency'], bins=30, ax=axes[1])
axes[1].set_title('Frequency')

sns.histplot(rfm['Monetary'], bins=30, ax=axes[2])
axes[2].set_title('Monetary')

plt.show()

```

```python
top_customers = rfm.sort_values('Monetary', ascending=False).head(10)
sns.barplot(x='CustomerID', y='Monetary', data=top_customers)
plt.xticks(rotation=45)
plt.show()
```

```python
counts = rfm['Churn Flag'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    counts,
    labels=counts.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title('Churn Flag Distribution')
plt.show()

```

```python
def rfm_segment(row):
    if row['R_Score'] >= 4 and row['F_Score'] >= 4:
        return 'Champions'
    elif row['R_Score'] >= 3 and row['F_Score'] >= 3:
        return 'Loyal'
    elif row['R_Score'] >= 4 and row['F_Score'] <= 2:
        return 'New'
    elif row['R_Score'] <= 2 and row['F_Score'] >= 3:
        return 'At Risk'
    else:
        return 'Lost'

rfm['RFM_Segment'] = rfm.apply(rfm_segment, axis=1)
```

```python
segment_revenue = rfm.groupby('RFM_Segment')['Monetary'].sum().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(x='RFM_Segment', y='Monetary', data=segment_revenue)
plt.title('Revenue by RFM Segment')
plt.xticks(rotation=45)
plt.show()
```
